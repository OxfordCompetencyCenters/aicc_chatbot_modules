## Exercise 1 - Rule-based vs LLM

Below is a tiny ELIZA-style rule-based responder. Run it with the sample inputs and inspect the replies — notice how it only works when the input matches one of the hard-coded patterns.

Then implement `llm_reply` to send the same inputs to the LLM, and compare.

```python
import re

def eliza_reply(user_input: str) -> str:
    """Very small rule-based responder."""
    user_input = user_input.strip().lower()
    if match := re.match(r"i'?m\s+(.+)", user_input):
        return f"Why are you {match.group(1)}?"
    if "hello" in user_input or "hi" in user_input:
        return "Hello. How are you feeling today?"
    return "I'm not sure I understand. Could you rephrase?"

sample_inputs = [
    "I'm sad",
    "Hi there",
    "Can you help me debug a Python script that crashes when I import pandas?",
]

for text in sample_inputs:
    print(f"> {text}")
    print(f"  ELIZA: {eliza_reply(text)}\n")
```
